```python
from sklearn.decomposition import PCA
pca=PCA(0.95)#0.95 bcz we are retaining 95% of variance
pca_X=pca.fit_transform(x)
pca_x
```

```python
pca_X.shape#dim reduced 64 to 29
```

```python
print("explained var ratio is\n{},\n\n components: {}".format(pca.explained_variance_ratio_,pca.n_components))
```

```python
from sklearn.model_selection import train_test_split
x_train_pca,x_test_pca,y_train_pca,y_test_pca=train_test_split(pca_X,y,test_size=0.3,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(x_train_pca,y_train_pca)
model.score(x_test_pca,y_test_pca)
```

```python
#0.14+0.13 we have 27% of imp features retained from reduction
```

```python
from sklearn.model_selection import train_test_split
x_train_pca,x_test_pca,y_train_pca,y_test_pca=train_test_split(pca_x,y,test_size=0.3,random_state=30)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(x_train_pca,y_train_pca)
print("{}%".format(model.score(x_test_pca,y_test_pca)*100))
```
